## 5. Create a function named prep_iris that accepts the untransformed iris data, and returns the data with the transformations above applied.

```python
def clean_data(df):
    df = df.drop(columns='species_id')
    df = df.rename(columns = {'species_name': 'species'})
    dummy_df = pd.get_dummies(df[['species']], dummy_na=False, drop_first=[True, True])
    df = pd.concat([df, dummy_df], axis=1)
    return df
```

```python
#remove measurement_id and species_id
iris = df_iris.drop(columns=['measurement_id', 'species_id'])
iris
```

```python
def train_validate_test_split(iris, target, seed=123):
    '''
    This function takes in a dataframe, the name of the target variable
    (for stratification purposes), and an integer for a setting a seed
    and splits the data into train, validate and test. 
    Test is 20% of the original dataset, validate is .30*.80= 24% of the 
    original dataset, and train is .70*.80= 56% of the original dataset. 
    The function returns, in this order, train, validate and test dataframes. 
    '''
    train_validate, test = train_test_split(iris, test_size=0.2, 
                                            random_state=seed, 
                                            stratify=iris[target])
    train, validate = train_test_split(train_validate, test_size=0.3, 
                                       random_state=seed,
                                       stratify=train_validate[target])
    return train, validate, test
```

```python
#split data into train, validate, test
train, validate, test = train_validate_test_split(iris, target='species_name')
train.head()
```
